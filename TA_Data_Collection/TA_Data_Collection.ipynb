## Data Collection

```python
import newspaper
from newspaper import article
```

```python
tribune = newspaper.build('https://tribune.com.pk/')
```

```python
for article in tribune.articles:
    try:
        article.download()
        article.parse()
        #article.nlp()
        print(article.title)
    except:
        continue
```

```python
import pandas as pd
News = pd.DataFrame()
print(News)
```

```python
for sentence in tribune.articles:
    try:
        sentence.download()
        sentence.parse()
        News = News.append({'sentence': sentence.title}, ignore_index=True)

        # export the dataframe to a CSV file
        News.to_csv('sentences.csv', index=False)

    except:
        continue
```
